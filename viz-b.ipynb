# Data Visulization for tips

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python

df = sns.load_dataset('tips')
df.head(5)
```

```python

sns.relplot(x='tip', y='total_bill', hue=df.tip.tolist(), data=df);
```

```python
tip_mean = df.groupby('tip', as_index=False).mean()
tip_mean
```

```python

sns.relplot(x='tip', y='total_bill',hue=tip_mean.tip.tolist(), data=tip_mean);
```

```python
sns.lmplot(x='tip', y='total_bill', data=tip_mean);
```

```python
tips_day = df.groupby('sex', as_index=False).sum()
tips_day
```

```python

sns.barplot(x='sex', y='tip', data=tips_day);
```
